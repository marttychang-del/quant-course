## 持股邏輯 → 量化因子

在下方表格中填入你的持股和買入邏輯，
寫出對應的量化代理指標。


```python
my_stocks = ["2330.TW", "2454.TW", "AAPL"]
data = []
for ticker in my_stocks:
    stock = yf.Ticker(ticker)
    hist = stock.history(period="1y")
    current_price = hist['Close'].iloc[-1]
    ma20 = hist['Close'].rolling(20).mean().iloc[-1]
    vol_20d = hist['Close'].pct_change().rolling(20).std().iloc[-1] * np.sqrt(252)
    data.append({
        'Ticker': ticker,
        'Price': round(current_price, 2),
        'MA20': round(ma20, 2),
        'Price/MA20': round(current_price / ma20, 3),
        'HV_20d': f"{vol_20d:.1%}",
        '買入邏輯': '（請自行填寫）',
        '量化因子': '（請自行填寫）'
    })
pd.DataFrame(data)
```
