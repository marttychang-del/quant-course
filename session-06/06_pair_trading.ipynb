# 第 6 堂：統計套利 — 配對交易


## 配對交易實驗

**自行修改 `pairs` 清單，測試不同的股票對。**

- Z-score > 2：Short A / Long B
- Z-score < -2：Long A / Short B
- |Z-score| < 1：無明顯偏離


```python
pairs = [
    ("2330.TW", "2454.TW"),
    ("2891.TW", "5880.TW"),
    ("AAPL", "MSFT"),
]
results = []
for a_ticker, b_ticker in pairs:
    try:
        a = yf.Ticker(a_ticker).history(period="1y")['Close']
        b = yf.Ticker(b_ticker).history(period="1y")['Close']
        aligned = pd.concat([a, b], axis=1).dropna()
        corr = aligned.iloc[:, 0].corr(aligned.iloc[:, 1])
        ratio = aligned.iloc[:, 0] / aligned.iloc[:, 1]
        z = (ratio.iloc[-1] - ratio.mean()) / ratio.std()
        results.append({"Pair": f"{a_ticker}/{b_ticker}", "Corr": round(corr, 4), "Z": round(z, 2)})
        print(f"{a_ticker}/{b_ticker}: corr={corr:.4f}, z={z:.2f}")
    except Exception as e:
        print(f"Error: {e}")
pd.DataFrame(results)
```
