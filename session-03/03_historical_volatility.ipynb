# 第 3 堂：基礎統計與金融時間序列 — 歷史波動率


## 練習：計算 N 日歷史波動率 (HV)

波動率是衡量**風險**最直覺的指標。
年度化 HV = 日波動率 × √252


```python
TICKER = "2330.TW"
df = yf.Ticker(TICKER).history(period="1y")[['Close']].copy()
log_returns = np.log(df['Close'] / df['Close'].shift(1)).dropna()
for n in [20, 60, 120]:
    hv = log_returns.rolling(window=n).std() * np.sqrt(252)
    print(f"{n}日HV（最新）：{hv.iloc[-1]:.2%}")
hv20 = log_returns.rolling(20).std() * np.sqrt(252)
hv60 = log_returns.rolling(60).std() * np.sqrt(252)
pd.DataFrame({'HV_20': hv20, 'HV_60': hv60}).plot(figsize=(12, 5))
plt.title(f'{TICKER} 年度波動率')
plt.ylabel('HV')
plt.show()
```
