## 練習：抓取股票歷史資料並繪圖

把 `TICKER` 換成你想看的股票：
- 台積電：`2330.TW`
- 聯發科：`2454.TW`
- Apple：`AAPL`
- Tesla：`TSLA`


```python
# 請修改這行，換成你想看的股票
TICKER = "2330.TW"
stock = yf.Ticker(TICKER)
df = stock.history(period="6mo")
print(f"資料筆數：{len(df)}")
df.head()
```

```python
df['Close'].plot(figsize=(12, 5), title=f'{TICKER} 收盤價')
plt.ylabel('價格')
plt.show()
```

```python
print("缺失值統計：")
print(df.isnull().sum())
df['Return'] = df['Close'].pct_change()
print(f"\n平均日報酬率：{df['Return'].mean():.4%}")
print(f"日波動率：{df['Return'].std():.4%}")
```
