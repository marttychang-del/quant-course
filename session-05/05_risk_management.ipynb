## Kelly Criterion + 2% 部位計算

**請根據你自己回測的結果修改下面三個數字。**


```python
# === 請修改這三個數字 ===
win_rate = 0.55
avg_win   = 0.10
avg_loss  = 0.05
kelly = (win_rate * avg_win - (1 - win_rate) * avg_loss) / avg_win
print(f"Kelly %（理論）：{kelly:.2%}")
print(f"建議實戰（半 Kelly）：{kelly/2:.2%}")
account_size = 1_000_000
risk_pct = 0.02
max_loss = account_size * risk_pct
entry_price = 100
stop_loss_price = 95
risk_per_share = entry_price - stop_loss_price
shares = int(max_loss / risk_per_share)
position_value = shares * entry_price
print(f"\n帳戶：{account_size:,.0f} 元")
print(f"單筆最大損失（2%）：{max_loss:,.0f} 元")
print(f"建議買入股數：{shares} 股")
print(f"總部位價值：{position_value:,.0f} 元（{position_value/account_size:.1%} 帳戶）")
```
